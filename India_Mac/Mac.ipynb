```python
data=pd.read_csv('Dataset\India_Menu.csv')
data.head()
```

Now let's look at the menu with the most energy

```python
data.groupby(['Menu Category'])['Energy (kCal)'].sum().sort_values(ascending=False)
```

```python
data.groupby(['Menu Category'])['Energy (kCal)'].sum().plot.barh()
plt.xlabel('Energy (kCal)')
```

most Total fat 

```python
data.groupby(['Menu Category'])['Total fat (g)'].sum().plot.barh()
plt.xlabel('Total fat (g)')
```

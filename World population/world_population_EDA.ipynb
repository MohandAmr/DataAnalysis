```python
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px
import missingno as msngo
from wordcloud import WordCloud
```

<h3 id="2" style="text-align: left; background-color: #128EB3; font-family:newtimeroman; color: white; padding: 14px; line-height: 1; border-radius:10px">üóëCleaning Data</h3>

```python
df=pd.read_csv('world_population.csv')
df.head()
```

```python
for col in df.columns:
    if 'Population' and '0' in col:
        df = df.rename(columns={col: col.split(' ')[0]})
        
df.head()
```

```python
df.drop(columns=['CCA3','Capital'],inplace=True)
```

```python
msngo.matrix(df)
```

```python
df.describe().T.sort_values(ascending = 0,by = "mean").style.background_gradient(cmap = "jet")
```

<h3 id="3" style="text-align: left; background-color: #128EB3; font-family:newtimeroman; color: white; padding: 14px; line-height: 1; border-radius:10px"> üìäplotting and visualization</h3>

```python
years=['1970','1980','1990','2000','2010','2015','2020','2022']
for x in years:
    fig = px.sunburst(df, path=['Continent', 'Country'], values=x,color_continuous_scale='jet',
                 template='plotly_dark', color=x,title=x)
    fig.show()
    fig = px.choropleth(df,
                     locations='Country',
                     locationmode='country names',
                     color=x,
                     color_continuous_scale='jet',
                     template='plotly_dark',
                     title = x)

    fig.show()

    fig = px.treemap(df, path=[px.Constant("world"), 'Continent', 'Country'], values=x,
                  color=x,template='plotly_dark',title = x,
                  color_continuous_scale='RdBu')
    fig.show()
```

```python
fig = px.choropleth(df,
                     locations='Country',
                     locationmode='country names',
                     color='Growth Rate',
                     color_continuous_scale='jet',
                     template='plotly_dark',
                     title = 'Growth Rate')

fig.show()

```

```python
i=[0,1]
for x in i :
    title="Top 10 Smallest Country" if(x) else "Top 10 Biggest Country"
    top_most_population = df.groupby(by = 'Country')['2022'].sum().sort_values(ascending=x).head(10)
    fig=px.bar(x=top_most_population.index,
            y=top_most_population.values,
            color=top_most_population.index,
            color_discrete_sequence=px.colors.qualitative.Safe,
            text=top_most_population.values,
            title=title,
            template= 'plotly_dark')
    fig.update_layout(
    xaxis_title="Countries",
    yaxis_title="Population",)
    fig.show()
```

```python
pop = df.melt(id_vars=['Country'], value_vars=['2020', '2010', '2000', '1990', '1980', '1970'], var_name='Year', value_name='Population')
pop = pop.sort_values('Year')
fig = px.choropleth(pop, 
              locations = 'Country',
              color="Population", 
              template='plotly_dark',
              animation_frame="Year",
              color_continuous_scale='IceFire',
              locationmode='country names',
              title = 'Growth Rate(Year 1970-2020)',
              height=600
             )
fig.show()
```

<h3 id ="4"style="text-align: left; background-color: #128EB3; font-family:newtimeroman; color: white; padding: 14px; line-height: 1; border-radius:10px">üåçWordCloud on Area Locality</h3>

```python
text = " ".join(subject_titles for subject_titles in df["Country"])

wordcloud = WordCloud().generate(text)


# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
